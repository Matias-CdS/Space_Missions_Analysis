```python
import pandas as pd
```

1. ¿Cuántos valores nulos encontrás en los datos? ¿Los puedes eliminar?


#No se encontrarón valores nulos en los datos.

#No es necesario eliminar valores.

```python
# Contar valores nulos por columna
print(datos.isnull().sum())

# Contar el total de valores nulos
print("Total de valores nulos:", datos.isnull().sum().sum())
```

2. ¿Cuántos valores incompletos encontrás en los datos? ¿Los puedes reemplazar?

#No existen valores vacíos.

#Se encontrarón valores incompletos 'Not Given' = total 2875 datos sin información.


```python
# Contar valores vacíos ("")
valores_vacios = (datos == "").sum().sum()

# Contar valores con "not given"
valores_not_given = (datos == "Not Given").sum().sum()

# Total de valores incompletos
total_incompletos = valores_vacios + valores_not_given

print(f"Valores vacíos: {valores_vacios}")
print(f"Valores con 'Not Given': {valores_not_given}")
print(f"Total de valores incompletos: {total_incompletos}")
```

#Datos reemplazados de 'Not Given' a 'Desconocido'.

```python
datos.replace("Not Given", "Desconocido", inplace=True)  # Not Given lo convierte en "Desconocido"
datos
```

3. ¿Podés eliminar columnas que no te aportan información? ¿Cuáles son? ¿Por qué las eliminarías?

5. La columna “listed_in” contiene diferentes valores separados por coma, ¿Podés crear una columna y quedarte con el primer valor?


#Creé una columna con nombre 'primer_valor' y nos quedamos con el primer valor. 

```python
nuevos_datos['primer_valor'] = nuevos_datos['listed_in'].str.split(',').str[0]
#.str.split(',') => Sirve para dividir la cadena en una lista usando la coma como separador.
#.str[0] => Sirve para obtener el primer elemento de la lista. 
print(nuevos_datos[['listed_in', 'primer_valor']].head()) #head => sirve para obtener los primeros 5 datos.
```
